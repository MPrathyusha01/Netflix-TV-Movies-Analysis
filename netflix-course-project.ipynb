## Introduction
This dataset consists of tv shows and movies available on Netflix as of 2019. The dataset is collected from Flixable which is a third-party Netflix search engine. The dataset contains 6234 titles (TV Series and Movies) available on Netflix. There are 12 columns related to each title:

* show_id - unique id of the title
* type - type of the title [TV Show, Movie]
* title - title name
* director - director of the movie
* cast - Leads of the title
* country - Country of origin
* date_added - date on which the title was added to Netflix
* release_year - year of release.
* rating - TV-rating
* duration - Runtime of the title (in minutes or seasons)
* listed_in - Genre
* description - short summary of the title

### Drop the null values

### Chage the Date- Time format

```python
netflix["date_added"] = pd.to_datetime(netflix['date_added'])
netflix['day_added']=netflix['date_added'].dt.day
netflix['month_added']=netflix['date_added'].dt.month
netflix['year_added']=netflix['date_added'].dt.year
netflix.head()
```

## Dividing the dataset with the type of shows

```python
movies_df = netflix_df[netflix_df['type'] == 'Movie']
tv_df = netflix_df[netflix_df['type'] == 'TV Show']
```

# Asking and Answering Questions using Visualization

##  What type of shows people watch more often

```python
sns.countplot(netflix_df['type'],alpha=0.8,palette='Set2')
plt.title('Type');
```

* By the above graph, we can conclude that there are `more movies viewrs` in netflix
* There are nearly more than 2000 vieres in movies than compared with TV shows

## What is the count of rating based on the category

```python
sns.countplot(x="year_added",data=netflix,palette='Paired');
```

* By anaysing the above graph, We can conclude that the number of movies views by the people as the time increases
*  **NOTE:** This data is taken in the mid 2020 so the 2020 analysis is incomplete

### On which day of the month, There are more number of movies or TV shows are released on Netflix

```python
netflix['day_added'].value_counts().plot.pie(figsize=(10,8))
plt.show()
```

* By the above piechart, we can conclude thet the more number of movies and tv shows are realesed on `first day` of the month

### On which month of the year, There are more number of Movies or TV shows releases in Netflix

```python
plt.figure(figsize = (35,10))
sns.countplot(x='month_added',data = netflix)
plt.title("Month of the Shows on Netflix")
```

* By the visualization, We can say that at the `start and the end of the month` , There are more movies and TV shows are released.

### What are the top 10 generes in the Netflix 

```python
plt.figure(figsize=(20,10))
netflix_df[netflix_df["type"]=="Movie"]["listed_in"].value_counts()[:10].plot(kind="barh",color="red")
plt.title("Top 10 Genres of Movies",size=18)
```

* We can easily say that `Documentation` has more number of views than any other generes

### Who is the director with most movies

```python
plt.figure(1, figsize=(20, 7))
plt.title("Director with most movies")
sns.countplot(x = "director", order=movies_df['director'].value_counts().index[0:10] ,data=movies_df,palette='Blues')
```

* By the above bar graph, we can say that the directer named `Raul Compos, Jan Suter` directer more movies

### Who is the director with most TV shows

```python
plt.figure(1, figsize=(20, 7))
plt.title("Director with most TV show")
sns.countplot(x = "director", order=tv_df['director'].value_counts().index[0:10] ,data=tv_df,palette='YlOrBr')
```

* By the above graph , we can conclude that the director named `Alastair Fothergill` directed more TV shows

```python
plt.figure(1, figsize=(20, 7))
sns.countplot(x = "country", order=movies_df['country'].value_counts().index[0:10] ,data=movies_df,palette='vlag')
```

* By the above graph we can conclude that `United states` make more number of movies in prime compare to other countries

## What are the top 25 oldest movies in netflix

```python
old_movies=movies_df[movies_df.release_year.isin(movies_df['release_year'].nsmallest(25))]
old_movies.sort_values(by='release_year')
```

* They are the `top 25 oldest movies` in Netflix

## What are the top 15 oldest TV shows in netflix

```python
old_tv_shows=tv_df[tv_df.release_year.isin(tv_df['release_year'].nsmallest(15))]
old_tv_shows.sort_values(by='release_year')
```

* These are the `Top 15 oldest movies` in Netflix

## What are the top 10 TV show generes in Netflix

```python
plt.figure(figsize=(20,10))
tv_df["listed_in"].value_counts()[:10].plot(kind="barh",color="blue")
plt.title("Top 10 Genres of Movies",size=18)
```

* They are the `top 10 TV shows` in the netflix
